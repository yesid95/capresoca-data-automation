```python
# Normalizar llaves para evitar fallos por espacios o tipos distintos
Df_SIE["tipo_documento"] = Df_SIE["tipo_documento"].astype(str).str.strip()
Df_SIE["numero_identificacion"] = Df_SIE["numero_identificacion"].astype(str).str.strip()

maestro_ADRES["TPS_IDN_ID"] = maestro_ADRES["TPS_IDN_ID"].astype(str).str.strip()
maestro_ADRES["HST_IDN_NUMERO_IDENTIFICACION"] = (
    maestro_ADRES["HST_IDN_NUMERO_IDENTIFICACION"]
    .astype(str)
    .str.strip()
)

# Columnas a traer desde maestro_ADRES
cols_adres = [
    "TPS_IDN_ID",
    "HST_IDN_NUMERO_IDENTIFICACION",
    "ENT_ID",
    "TPS_EST_AFL_ID",
    "CRUCE_BDEX_RNEC",
    "MARCASISBENIV+MARCASISBENIII_2",
]

# Cruce
Df_SIE = Df_SIE.merge(
    maestro_ADRES[cols_adres],
    how="left",
    left_on=["tipo_documento", "numero_identificacion"],
    right_on=["TPS_IDN_ID", "HST_IDN_NUMERO_IDENTIFICACION"],
    indicator=True
)

# Conteo de cruce
conteo_cruce_adres = Df_SIE["_merge"].value_counts()

print("Registros que cruzaron con maestro_ADRES:", conteo_cruce_adres.get("both", 0))
print("Registros que no cruzaron con maestro_ADRES:", conteo_cruce_adres.get("left_only", 0))

# Opcional: eliminar columnas llave duplicadas y columna auxiliar
Df_SIE = Df_SIE.drop(
    columns=["TPS_IDN_ID", "HST_IDN_NUMERO_IDENTIFICACION", "_merge"]
)
```
