```python
import os
import pandas as pd
import glob
import chardet
from datetime import datetime
```

# Maestros ADRES 

```python
# Ruta de la carpeta
folder = r"C:\Users\osmarrincon\OneDrive - uniminuto.edu\Capresoca\AlmostClear\Procesos BDUA\Subsidiados\Maestro\MS\2025-02"

# Listar archivos .txt
files = glob.glob(os.path.join(folder, "*.TXT"))

data = []
for file in files:
    filename = os.path.basename(file)
    # Extraer fecha del nombre (últimos 12 a 4 caracteres antes de .TXT)
    date_str = filename[-12:-4]
    fecha = f"{date_str[:2]}/{date_str[2:4]}/{date_str[4:]}"
    
    # Crear objeto datetime para ordenamiento
    fecha_datetime = datetime.strptime(fecha, "%d/%m/%Y")
    
    # Detectar encoding
    with open(file, 'rb') as f:
        raw = f.read(10000)
        result = chardet.detect(raw)
        encoding = result['encoding']
    
    # Detectar separador y contar columnas
    with open(file, 'r', encoding=encoding, errors='ignore') as f:
        first_line = f.readline().strip()
        if ',' in first_line:
            sep = ','
            num_columnas = len(first_line.split(','))
        elif '|' in first_line:
            sep = '|'
            num_columnas = len(first_line.split('|'))
        elif ';' in first_line:
            sep = ';'
            num_columnas = len(first_line.split(';'))
        else:
            sep = 'Desconocido'
            num_columnas = 0
    
    data.append({
        'Archivo': filename,
        'Fecha': fecha,
        'Fecha_datetime': fecha_datetime,
        'Encoding': encoding,
        'Separador': sep,
        'Num_Columnas': num_columnas
    })

# Crear DataFrame y ordenar por fecha
df_archivos = pd.DataFrame(data)
df_archivos = df_archivos.sort_values('Fecha_datetime')

# Eliminar la columna auxiliar de datetime y resetear índice
df_archivos = df_archivos.drop('Fecha_datetime', axis=1).reset_index(drop=True)

df_archivos
```
