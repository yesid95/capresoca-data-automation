# Modulo

```python
import os
import pandas as pd
import numpy as np
from datetime import datetime
```

# Rutas y dataframes

```python
R_R1_contruir = r"C:\Users\osmarrincon\Downloads\txt\validar-R1.txt"
R_R1 = r"C:\Users\osmarrincon\Downloads\txt\R1.txt"

R_BDUA = r"C:\Users\osmarrincon\OneDrive - 891856000_CAPRESOCA E P S\Capresoca\AlmostClear\Procesos BDUA\Contributivo\Procesos BDUA EPS\AUTOMATICO R1\All-AUTO-R1.txt"
```

```python
df_R1_Validar = pd.read_csv(R_R1_contruir, dtype=str, sep=',', header=0, encoding='ANSI')
df_R1_BDUA = pd.read_csv(R_BDUA, dtype=str, sep=',', header=0, encoding='ANSI')
```

# Depurar R1 BDUA

```python
# Convertir la columna Fecha_Efectiva a formato datetime
df_R1_BDUA['Fecha_Efectiva'] = pd.to_datetime(df_R1_BDUA['Fecha_Efectiva'], format='%d/%m/%Y')

# Ordenar por fecha efectiva en orden descendente
df_R1_BDUA_sorted = df_R1_BDUA.sort_values('Fecha_Efectiva', ascending=False)

# Eliminar duplicados manteniendo el primer registro (el más reciente)
df_R1_BDUA_depurado = df_R1_BDUA_sorted.drop_duplicates(
    subset=['TPS_IDN_ID', 'HST_IDN_NUMERO_IDENTIFICACION'], 
    keep='first'
)

# Opcional: volver a ordenar el índice
df_R1_BDUA_depurado = df_R1_BDUA_depurado.reset_index(drop=True)

print(f"Registros originales: {len(df_R1_BDUA)}")
print(f"Registros únicos (más recientes): {len(df_R1_BDUA_depurado)}")
```

```python
# Imprimir número de registros antes del proceso
print(f"Registros en df_R1_BDUA_depurado antes del filtro: {len(df_R1_BDUA_depurado)}")
print(f"Registros en df_R1_Validar: {len(df_R1_Validar)}")

# Filtrar df_R1_BDUA_depurado para dejar solo los registros que están en df_R1_Validar
df_R1_BDUA_filtrado = df_R1_BDUA_depurado.merge(
    df_R1_Validar[['TPS_IDN_ID', 'HST_IDN_NUMERO_IDENTIFICACION']],
    on=['TPS_IDN_ID', 'HST_IDN_NUMERO_IDENTIFICACION'],
    how='inner'
)

# Imprimir número de registros después del proceso
print(f"Registros en df_R1_BDUA_filtrado después del filtro: {len(df_R1_BDUA_filtrado)}")
```

```python
# Guardar df_R1_BDUA_filtrado en archivo txt
df_R1_BDUA_filtrado.to_csv(R_R1, sep=',', index=False, encoding='ANSI')

print(f"Archivo guardado exitosamente en: {R_R1}")
print(f"Total de registros guardados: {len(df_R1_BDUA_filtrado)}")
```
