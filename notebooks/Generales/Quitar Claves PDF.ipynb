```python
from pathlib import Path
from pypdf import PdfReader, PdfWriter

# =========================
# CONFIGURACIÓN (editar aquí en cada ejecución)
PDF_PATH = r"C:\Users\crist\Downloads\certificadoBasicoBancaMovil.pdf"
PDF_PASSWORD = "1118573708"
# Opcional: cambia el nombre de salida; si es None, usa "<nombre>_sin_clave.pdf"
OUTPUT_PATH = None
# =========================


def quitar_clave_pdf(pdf_path: str, password: str, output_path: str | None = None) -> Path:
    input_file = Path(pdf_path)

    if not input_file.exists():
        raise FileNotFoundError(f"No existe el archivo: {input_file}")

    if output_path is None:
        output_file = input_file.with_name(f"{input_file.stem}_sin_clave{input_file.suffix}")
    else:
        output_file = Path(output_path)

    reader = PdfReader(str(input_file))

    if reader.is_encrypted:
        
        ok = reader.decrypt(password)
        if ok == 0:
            raise ValueError("La contraseña es incorrecta o no se pudo descifrar el PDF.")

    writer = PdfWriter()
    for page in reader.pages:
        writer.add_page(page)

    with open(output_file, "wb") as f:
        writer.write(f)

    return output_file

if __name__ == "__main__":
    salida = quitar_clave_pdf(PDF_PATH, PDF_PASSWORD, OUTPUT_PATH)
    print(f"PDF sin clave guardado en: {salida}")
```
